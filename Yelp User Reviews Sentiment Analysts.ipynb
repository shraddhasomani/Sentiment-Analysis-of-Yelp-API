```python
# Loop through each review run the sentiment analysis on each sentence
polarity = []
subjectivity = []
for sen in df60['Reviews']:
    polarity.append(TextBlob(sen).sentiment.polarity)
    subjectivity.append(TextBlob(sen).sentiment.subjectivity)
```
